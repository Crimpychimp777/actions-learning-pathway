```python
def count_books_with_high_rating(dataframe, min_rating=4):
    """
    Counts the number of books in a DataFrame with an average rating greater than a specified minimum.

    Args:
        dataframe (pd.DataFrame): The input DataFrame containing book data.
        min_rating (int or float): The minimum average rating to filter by.

    Returns:
        int: The number of books with an average rating greater than `min_rating`.
    """
    high_rated_books = dataframe[dataframe['average_rating'] > min_rating]
    return len(high_rated_books)

# Use the function to find the number of books with an average rating greater than 4
high_rating_count = count_books_with_high_rating(df, min_rating=4)
print(f"Number of books with an average rating greater than 4: {high_rating_count}")
```
